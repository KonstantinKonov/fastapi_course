```python
import asyncio
import aiohttp
```

```python
async def get_data(i: int, endpoint: str):
    url = f"http://127.0.0.1:8000/{endpoint}/{i}"
    async with aiohttp.ClientSession() as session:
        async with session.get(url) as resp:
            print(f"finished execution: {i}")
```

```python
await get_data(1, "sync")
```

```python
await get_data(1, "async")
```

```python
await asyncio.gather(
    *[get_data(i, "async") for i in range(3000)]
)
```

```python
await asyncio.gather(
    *[get_data(i, "sync") for i in range(3000)]
)
```
